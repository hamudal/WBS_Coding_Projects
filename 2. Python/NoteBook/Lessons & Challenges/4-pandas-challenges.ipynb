###### **Exercise 7:**
change the column names so that there are no names with spaces or weird special characters:

```python
cars.columns = cars.columns.str.replace(' |/', '',regex=True)
```

###### **Exercise 10:**
did cars consume more gas in the eighties?

```python
cars[cars['Year']<1990]['FuelBarrelsYear'].mean()
```

```python
cars["Year"].between(1979, 1990)
```

```python
cars['eighties'] = cars["Year"].between(1979, 1990)
cars.groupby('eighties').mean()
```

```python
cars.groupby('Year')['FuelBarrelsYear'].mean()
```

```python
cars['decade'] = pd.cut(cars.Year, [1980,1990,2000,2010,2020], include_lowest=True, right=False, labels=['1980s','1990s','2000s','2010s'])
cars.groupby('decade')['FuelBarrelsYear'].mean()
```

###### **Exercise 12:**
read the `car_brands.csv` data:

```python
# car_brands.csv 'https://drive.google.com/file/d/1jLSArDeSsySULKiRtMNV3V3RKuWAUKbD/view?usp=sharing'

url = 'https://drive.google.com/file/d/1jLSArDeSsySULKiRtMNV3V3RKuWAUKbD/view?usp=sharing' # car_brands.csv
path = 'https://drive.google.com/uc?export=download&id='+url.split('/')[-2]
car_brands = pd.read_csv(path)
car_brands
```

###### **Exercise 13:**
join the cars dataframe with the car brands dataframe.

```python
cars.rename(columns={'Make':'brand'},inplace=True)
cars.join(car_brands.set_index('brand'),on='brand')
#cars
```

```python
cars.merge(car_brands,how='inner',left_on='Make',right_on='brand').drop("brand",axis=1)
```
