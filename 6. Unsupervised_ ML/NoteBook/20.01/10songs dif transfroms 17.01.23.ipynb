```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import pairwise_distances
```

```python
df_audio_features = pd.read_csv("df_audio_features_10.csv",
                                index_col=["song_name", "artist"])
df_audio_features.drop(['id','html'], axis=1, inplace=True)
df_audio_features
```

```python
# pip install scikit-learn  --upgrade
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
my_scaler_mm = MinMaxScaler().set_output(transform='pandas')
min_max_df = my_scaler_mm.fit_transform(df_audio_features)
min_max_df
```

```python
pd.DataFrame({
    'original': df_audio_features.iloc[:,-1], 
    'min_max_scale': min_max_df.iloc[:,-1]
}).sort_values(by='original')
```

```python
# Stating that we want two plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Plotting the two plots
sns.histplot(data=df_audio_features.iloc[:,3], bins=10, kde=True, ax=ax1);
sns.histplot(min_max_df.iloc[:,3], bins=10, kde=True, ax=ax2);

# Adding titles to the plots
ax1.set_title('Distribution of music without scaling')
ax2.set_title('Distribution of music with MinMax scaling')

plt.show()
```

```python
# Making the DataFrame for the MinMaxed heatmap
with_music_minmax_distances = pd.DataFrame(pairwise_distances(min_max_df), index=df_audio_features.index, columns=df_audio_features.index)

# Stating that we want two plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Plotting the two plots
sns.heatmap(df_audio_features, ax=ax1, linewidths=.2);
sns.heatmap(with_music_minmax_distances, ax=ax2, linewidths=.2);

# Adding titles to the plots
ax1.set_title('Including music - everything NOT in the same 1-10 scaling')
ax2.set_title('Including music - everything MinMax scaled')

plt.show()
```

Standard Scaler

```python
# Import Standard Scaler
from sklearn.preprocessing import StandardScaler

# Create DataFrame of Standar scaled values
my_scaler_st = StandardScaler().set_output(transform='pandas')
standard_df = my_scaler_st.fit_transform(df_audio_features)

# Display age column of original DataFrame and age column of MinMaStandard scaledxed DataFrame to compare
pd.DataFrame({
    'original': df_audio_features.iloc[:,-1], 
    'standard_scale': standard_df.iloc[:,-1]
}).sort_values(by='original')
```

```python
# Stating that we want two plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Plotting the two plots
sns.histplot(data=df_audio_features.iloc[:,-1], bins=10, kde=True, ax=ax1);
sns.histplot(standard_df.iloc[:,-1], bins=10, kde=True, ax=ax2);

# Adding titles to the plots
ax1.set_title('Distribution of music without scaling')
ax2.set_title('Distribution of music with Standard scaling')

plt.show()
```

```python
# Making the DataFrame for the Standard heatmap
with_age_standard_distances = pd.DataFrame(pairwise_distances(standard_df), index=df_audio_features.index, columns=df_audio_features.index)

# Stating that we want two plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

# Plotting the two plots
sns.heatmap(df_audio_features, ax=ax1, linewidths=.2);
sns.heatmap(with_age_standard_distances, ax=ax2, linewidths=.2);

# Adding titles to the plots
ax1.set_title('Including music - everything NOT in the same 1-10 scaling')
ax2.set_title('Including music - everything Standard scaled')

plt.show()
```
