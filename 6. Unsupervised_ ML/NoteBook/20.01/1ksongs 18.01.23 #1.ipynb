```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
```

```python
df_audio_features_1k = pd.read_csv("df_audio_features_1000.csv",
                                index_col=["name", "artist"])
df_audio_features_1k.drop(['id','html','type', 'time_signature'], axis=1, inplace=True)
df_audio_features_1k.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(data=df_audio_features_1k, x='energy', y='danceability')
plt.title('Music')
plt.show()
```

```python
# Applying KMean on it:
kmean_test = df_audio_features_1k[['energy','danceability']]
df_audio_features_1k_scaled = KMeans(n_clusters=5)
df_audio_features_1k_scaled.fit(kmean_test)
df_audio_features_1k_scaled.cluster_centers_
# Making another df out of it:
model_centroids = pd.DataFrame(df_audio_features_1k_scaled.cluster_centers_)
model_centroids.columns = ['energy','danceability']
model_centroids
```

```python
fig, ax = plt.subplots(figsize=(10, 8))

plt.title('Comparing our centroids and our dataset')
plt.xlabel('danceability')
plt.ylabel('energy')
sns.scatterplot(data=kmean_test, x='danceability', y='energy')
sns.scatterplot(data=model_centroids, x='danceability', y='energy', color='red', s=250)
plt.show()
```
